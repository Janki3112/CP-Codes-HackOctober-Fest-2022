## Load Data

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney.csv')
df
```

##**Data Preperation**


###Data seperation as X and Y

```python
y = df['measured log(solubility:mol/L)']
y
```

```python
X = df.drop('measured log(solubility:mol/L)', axis=1)
X
```

###Data Spitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

## **Model Building**

##**Linear Regression**


###**Training the Model**

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Assuming X_train is a pandas DataFrame, identifying columns with string values
string_columns = X_train.select_dtypes(include=['object']).columns

# Removing string columns which are not relevant
X_train_numeric = X_train.drop(columns=string_columns)

#fitting the model again with the modified data
lr = LinearRegression()
lr.fit(X_train_numeric, y_train)
```

```python
X_train_numeric
```

```python
y_test
```

###**Applying the Model to make a prediction**

```python
# Selecting the numeric column for prediction
X_test_numeric = X_test[['ESOL predicted log(solubility:mol/L)']]

# Now prediction using the numeric data
y_lr_test_pred = lr.predict(X_test_numeric)

print(y_lr_test_pred)
```

###**Evaluate the Model**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE Train:' , lr_train_mse)
print('LR R2 Train:' , lr_train_r2)
print('LR MSE Test:' , lr_test_mse)
print('LR R2 Test:' , lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['method', 'Traning MSE', 'Training R2', 'Testing MSE', 'Testing R2']
```

```python
lr_results
```

##**Random Forest**

###**Traning the Model**

```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd

rf = RandomForestRegressor(max_depth=2, random_state=100)

#fitting the model again with the modified data
rf.fit(X_train_numeric, y_train)
RandomForestRegressor(max_depth=2, random_state=100)
```

###**Applying the Model to make a prediction**

```python
# Selecting the numeric column for prediction
y_rf_train_pred = rf.predict(X_train_numeric)

# Now prediction using the numeric data
y_rf_test_pred = rf.predict(X_test_numeric)

print(y_rf_test_pred)
```

```python
y_rf_test_pred = rf.predict(X_test_numeric)
```

###**Evaluate the Model**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print('RF MSE Train:' , rf_train_mse)
print('RF R2 Train:' , rf_train_r2)
print('RF MSE Test:' , rf_test_mse)
print('RF R2 Test:' , rf_test_r2)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['method', 'Traning MSE', 'Training R2', 'Testing MSE', 'Testing R2']
```

```python
rf_results
```

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```
